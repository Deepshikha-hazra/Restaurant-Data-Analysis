### **Loading Dataset**

```python
#Importing all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/content/Restaurant Dataset.csv")
```

```python
df.head()
```

### **Analyze the distribution of aggregate ratings and determine the most common rating range.**

```python
# Analyze the distribution of aggregate

plt.figure(figsize=(10,5))
plt.hist(df['Aggregate rating'],edgecolor = 'black')
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurant')
plt.show()
```

```python
# Determine the most common rating:
most_common_rating_range = df['Aggregate rating'].apply(lambda x:round(x/2)*2).mode()[0]
print("Most common rating range:",most_common_rating_range)
```

### **Calculate the average number of votes received by restaurants.**

```python
print("Average number of vote received by restaurant:", round(df['Votes'].mean(),2))
```
