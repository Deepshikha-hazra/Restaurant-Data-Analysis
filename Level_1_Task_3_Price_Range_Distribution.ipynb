### **Loading Dataset**

```python
# Importing all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/content/Restaurant Dataset.csv")
```

```python
df.head()
```

### **Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.**

```python
price_range_count =df['Price range'].value_counts().sort_index()
```

```python
total_restaurant = len(df)
percentage_per_price_range =(price_range_count / total_restaurant) * 100
```

```python
plt.bar(price_range_count.index, price_range_count.values)
plt.xlabel('Price range')
plt.ylabel('count')
plt.title('Distribution of Price Range')
plt.show()
```

### **Calculate the percentage of restaurants in each price range category.**

```python
print("Percentage of restaurant in each price range categories:")
print(percentage_per_price_range)
```
