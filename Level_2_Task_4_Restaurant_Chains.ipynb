### **Loading Dataset**

```python
# Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/content/Restaurant Dataset.csv')
```

### **Identify if there are any restaurant chains present in the dataset.**

```python
# Total count of each restaurant
restaurant_counts = df['Restaurant Name'].value_counts()
restaurant_counts
```

```python
# Identify Restaurant chains
restaurant_chains = restaurant_counts[restaurant_counts > 1].index.tolist()
print("Restaurant Chains:",restaurant_chains)
```

### **Analyze the ratings and popularity of different restaurant chains.**

```python
# Filter out the data for restaurant chains only:
restaurant_chaines_data = df.loc[df['Restaurant Name'].isin(restaurant_chains)]
```

```python
# Calculate average rating and popularity(i.e votes)for each restaurant chains:
restaurant_chains_avg_rating = restaurant_chaines_data.groupby('Restaurant Name')['Aggregate rating'].mean()
chains_total_votes= restaurant_chaines_data.groupby('Restaurant Name')['Votes'].sum()
```

```python
# Create a dataframe and stare the claculate data:
chain_information = pd.DataFrame({'Average rating' : restaurant_chains_avg_rating, 'Total votes': chains_total_votes})
print("Restaurant chain ratings and popularities")
chain_information = chain_information.sort_values('Average rating', ascending=False)
chain_information
```
